```python
from sentence_transformers import SentenceTransformer, util
from transformers import pipeline

# Embedding model to represent sentences/documents
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Optional: Load a fact-checking classifier (e.g., from Hugging Face hub)
fact_check_classifier = pipeline('text-classification', model='facebook/bart-large-mnli')
```

```python
# Example source document and generated text
source_text = "Albert Einstein was a theoretical physicist born in Germany in 1879."
generated_text = "Albert Einstein was born in 1689 in the United States."

# Compute embeddings
source_emb = embedding_model.encode(source_text, convert_to_tensor=True)
generated_emb = embedding_model.encode(generated_text, convert_to_tensor=True)

# Compute cosine similarity
similarity = util.cos_sim(source_emb, generated_emb).item()
print(f"Cosine similarity between source and generated text: {similarity:.3f}")

# Threshold example: if similarity < 0.7, flag as possible hallucination
if similarity < 0.7:
    print("Warning: Possible hallucination detected!")
else:
    print("Generated text is likely consistent with source.")
```
